```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
data_LIT = pd.read_excel("statistiky.xlsx", sheet_name="ZapasyLIT", index_col=3)
# Data in Excel file - it contains hockey matches statistics of HC Verva Litvínov from Czech Republic, they are matches from season 2018/2019
```

```python
souperi = data_LIT["Soupeř"].unique()
print(souperi)
# List of all Litvínov´s opponents
```

```python
data_LIT = data_LIT.fillna(0)
# Deleting of raws with some empty cell
```

```python
data_LIT[["Góly LIT první třetina", "Góly soupeř první třetina"]] = \
    data_LIT["1. třetina"].str.split(":", expand=True)

data_LIT["Góly LIT první třetina"] = \
    data_LIT["Góly LIT první třetina"].astype(int)

data_LIT["Góly soupeř první třetina"] = \
    data_LIT["Góly soupeř první třetina"].astype(int)

data_LIT[["Góly LIT druhá třetina", "Góly soupeř druhá třetina"]] = \
    data_LIT["2. třetina"].str.split(":", expand=True)

data_LIT["Góly LIT druhá třetina"] = \
    data_LIT["Góly LIT druhá třetina"].astype(int)

data_LIT["Góly soupeř druhá třetina"] = \
    data_LIT["Góly soupeř druhá třetina"].astype(int)

data_LIT[["Góly LIT třetí třetina", "Góly soupeř třetí třetina"]] = \
    data_LIT["3. třetina"].str.split(":", expand=True)

data_LIT["Góly LIT třetí třetina"] = \
    data_LIT["Góly LIT třetí třetina"].astype(int)

data_LIT["Góly soupeř třetí třetina"] = \
    data_LIT["Góly soupeř třetí třetina"].astype(int)

data_LIT["Góly LIT"] = \
    data_LIT["Góly LIT"].astype(int)

data_LIT["Góly soupeř"] = \
    data_LIT["Góly soupeř"].astype(int)

data_LIT["Rozdíl skóre"] = (data_LIT["Góly LIT"] - data_LIT["Góly soupeř"])
# Creating of calculated columns - goals for each match period
```

```python
data_LIT.to_csv("litvinov_matches.csv", index=False, header=False, encoding="utf-8")
# Saving of data to csv file
```

```python
umisteni_prumer = data_LIT.groupby("Sezóna")["Umístění LIT"].mean().sort_values(ascending=True)
print(umisteni_prumer)
# Average final standing of Litvínov in each regular season
```
